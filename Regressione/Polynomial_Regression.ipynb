# Linear Regression for Polynomial Regression Tasks

## Degree r=2

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

np.random.seed(14)

# size del training set
N = 500

# genero i punti: asse x 
x = np.random.exponential(scale=1.0, size=(N,1))
print(x.shape)

# genero i punti: asse y (target)
y = 2 * x**2 + -15*x + 1 + np.random.randn(N, 1)
print(y.shape)

# genero x_0=1, necessario per considerare il termine di bias
x_0 = np.ones((N, 1))
print(x_0.shape)

# creo la design matrix X includendo x_0
X = np.hstack((x_0, x, x**2))
print(X.shape)

# calcolo modello w = (X^T X)^-1 X^T y
w = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))

print(w)

plt.figure()
plt.scatter(x, y, marker='.')

xmin = x.min()
xmax = x.max()
x_grid = np.linspace(xmin, xmax, 100)

plt.plot(x_grid, w[0] + x_grid * w[1] + x_grid**2 * w[2], color='red')
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(np.hstack((x, x**2)), y) #passiamo il training data, non la design matrix
print(f'Intercept: {lr_model.intercept_}, coeff: {lr_model.coef_}')
```

## Degree r=4

```python
x = np.random.uniform(0,4,size=(N,1))
y =x**4 -5 * x**3 + 5 * x**2 +5*x + -6 + np.random.randn(N, 1)
x_0 = np.ones((N, 1))
X = np.hstack((x_0, x, x**2, x**3, x**4))
w = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
print(w)

plt.figure()
plt.scatter(x, y, marker='.')
x_grid = np.linspace(0, 4, 100)
plt.plot(x_grid, w[0] + x_grid * w[1] + x_grid**2 * w[2] + x_grid**3 * w[3] + x_grid**4 * w[4], color='red')
```
