# Linkage-based clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

#example from slides we have seen in class (here with coordinates)
X = [[1,10],[1,1],[1.5,1.5],[5,2],[5,3]] 
#X = [[1,10],[1,1],[1.5,1.5],[5,2],[1,11]]
#point a = index 0
#point b = index 1
#point c = index 2
#point d = index 3
#point e = index 4

Z = linkage(X, method='single', metric='euclidean')

plt.figure(figsize=(25, 10))
dendrogram(Z)

plt.show()
```
