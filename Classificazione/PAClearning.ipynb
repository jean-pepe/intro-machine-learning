```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

def samplesize (epsilon, hsize, delta):
    return np.log(hsize/delta)/epsilon

delta= 0.1
epsilon = np.linspace(0.01, 0.1, 100)
hsize = np.array([i for i in range(1000,100000,1000)])

X, Y = np.meshgrid(epsilon, hsize)
print(X)
print(Y)
Z = samplesize(X,Y,delta)

fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='cool')
ax.set_title("Sample size m", fontsize = 13)
ax.set_xlabel('epsilon', fontsize = 11)
ax.set_ylabel('|H|', fontsize = 11)
ax.set_zlabel('m', fontsize = 11)
plt.show()
```
